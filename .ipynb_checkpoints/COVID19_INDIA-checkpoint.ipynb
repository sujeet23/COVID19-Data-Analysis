```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
C19_df = pd.read_csv('covid_19_india.csv')
```

# Data pre-processing

## 1) Dropping data and changing data type

```python
C19_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace = True, axis =1)
```

```python
C19_df.head(7)
```

changing date datatype

```python
C19_df["Date"]= pd.to_datetime(C19_df["Date"])
```

```python
C19_df.info()
```

```python
C19VAC_df = pd.read_csv('covid19_vaccine_statewise.csv')
```

```python
C19VAC_df.head(10)
```

```python
C19VAC_df.info()
```

## 2) Calculating Active cases

```python
C19_df["Active_cases"] = C19_df["Confirmed"] -(C19_df["Cured"]+C19_df["Deaths"])
```

```python
C19_df.tail()
```

## 3) Creating pivot table and sorting data

```python
C19_State = pd.pivot_table(C19_df, values = ["Confirmed","Deaths", "Cured"], index = "State/UnionTerritory", aggfunc = max)
```

```python
C19_State["Recovery Rate"] = C19_State["Cured"]*100/C19_State["Confirmed"]
```

```python
C19_State["Mortality Rate"] = C19_State["Deaths"]*100 /C19_State["Confirmed"]
```

```python
C19_State = C19_State.sort_values(by = "Confirmed", ascending = False)
```

```python
C19_State
```

## 4) Grouping

```python
#top 10 based on active case

C19_active = C19_df.groupby(by = 'State/UnionTerritory').max()[['Active_cases', 'Date']].sort_values(by=['Active_cases'], ascending = False).reset_index()
```

# Data Visualisation

## 1)Top 10 States Based on Active Cases

```python
fig = plt.figure(figsize =(16,9))

plt.title("Top 10 States Based on Active Cases", size = 24)

ax = sns.barplot(data= C19_active.iloc[:10], y = "Active_cases", x = "State/UnionTerritory", linewidth = 1.5, edgecolor = 'blue')

plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

## 2) Top 10 States with Most Deaths

```python
# Top 10 States with highest death 

C19_deaths = C19_df.groupby(by = 'State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by = ['Deaths'], ascending =False).reset_index()

fig = plt.figure(figsize=(16, 9))

plt.title("Top 10 States with Most Deaths", size = 24)

ax = sns.barplot(data = C19_deaths.iloc[:10], y= "Deaths", x = "State/UnionTerritory", linewidth = 2, edgecolor = 'red')

plt.xlabel("States")
plt.ylabel("Total Death cases")
plt.show()
```

## 3) Top 5 Affected states in India

```python
# Growth trend

fig = plt.figure(figsize=(16,9))

ax = sns.lineplot(data = C19_df[C19_df["State/UnionTerritory"].isin(["Maharashtra", "Delhi", "Gujarat", 'Tamil Nadu', 'Uttar Pradesh'])], x = "Date", y = "Active_cases", hue = "State/UnionTerritory")

ax.set_title("Top 5 Affected states in India",  size = 16)
```
